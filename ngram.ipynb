```python
from torch.utils.data import DataLoader
from moses import CharVocab, StringDataset
from moses.baselines.ngram import NGram
import moses

import rdkit.Chem as chem

train = moses.get_dataset('train')
```

```python
model = NGram(verbose=True)
model.fit(train)
```

```python
import pandas as pd
from io import StringIO
import sys
chem.WrapLogs()
sio = sys.stderr = StringIO()

total_per_n = 1_000_000
save_every = 25_000

with open('ngram_smiles.csv', 'w') as fh:
    fh.write(f"n,smiles,valid\n")


for n in range(2, 9):
    for i in range(total_per_n // save_every):
        vals = []
        for _ in range(save_every):
            smiles = model.generate_one(context_len=n)
            mol = chem.MolFromSmiles(smiles)
            if mol is None:
                sio = sys.stderr = StringIO()
            vals.append((smiles, mol is not None))
        print(f"Saving for n={n}, {i + 1}/{total_per_n // save_every}")
        with open('ngram_smiles.csv', 'a') as fh:
            for smiles, valid in vals:
                fh.write(f"{n},{smiles},{valid}\n")
```
